```python
for each in os.listdir(stopwords_path):
    each = stopwords_path + '/' + each
    with open(each,'r') as f:
        for line in f:
            for word in re.findall(r'\w+', line):
                stopwords.add(word) #We want all stopwords, in the set we made earlier.
```
